```python
df_arret_ligne = df_arret_lignes.drop_duplicates(subset=["shortName"], keep="first")
print("\nAprès suppression des doublons :")
df_arret_ligne
```

```python
dim_transport = df_arret_ligne.groupby("mode").agg(
    code_insee=('Code_insee', 'first'),  # Get the first 'nom_COMMUNE' in each group
    count=('mode', 'size')  # Count the occurrences of each 'Code_insee'
).reset_index()

dim_transport.insert(0, 'id_tp', ['P' + str(i) for i in range(1, len(dim_transport) + 1)])
```

```python
df_pollution = pd.concat([df_pollutionO3, df_pollutionNO2], ignore_index=True)
df_pollution.head(5)
```

```python
df_pollution.dtypes
```

```python

# Step 1: Group by the "Polluant" column and count the occurrences
dim_polluant = df_pollution.groupby("Polluant").size().reset_index(name="count")

# Step 2: Insert an "id_pol" column with custom IDs
dim_polluant.insert(0, 'id_pol', ['P' + str(i) for i in range(1, len(dim_polluant) + 1)])

dim_polluant = dim_polluant.drop(
    columns=[
        "count"
    ])
```

```python
dim_polluant.head()
```

```python
df_pollution["Date de début"]=pd.to_datetime(df_pollution['Date de début']).dt.normalize()

dim_time = df_pollution[["Date de début","Polluant"]]
dim_time["id_date"] = dim_time["Polluant"] + "_" + dim_time["Date de début"].astype(str)
dim_time["month"] = dim_time["Date de début"].dt.month
dim_time["year"] = dim_time["Date de début"].dt.year
dim_time["quarter"] = dim_time["Date de début"].dt.quarter
```

```python
dim_commune = df_arret_lignes

dim_commune = dim_commune.groupby("Code_insee").agg(
    nom_commune=('Nom_commune', 'first'),  # Get the first 'nom_COMMUNE' in each group
    count=('Code_insee', 'size')  # Count the occurrences of each 'Code_insee'
).reset_index()

dim_commune.insert(0, 'id_comm', ['C' + str(i) for i in range(1, len(dim_commune) + 1)])

dim_commune = dim_commune.drop(columns=["count"])

# Display the result
print(dim_commune)
```
